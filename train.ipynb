```python
import pandas as pd
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
try:
    train_df = pd.read_csv('training.csv')
    X_train = train_df['text']
    y_train = train_df['label']
except FileNotFoundError:
    print("Error: training.csv not found. Please run the generator script first.")
    exit()
```

```python
text_clf = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words='english', max_features=5000)),
    ('clf', MLPClassifier(
        hidden_layer_sizes=(100, 50), 
        max_iter=500, 
        activation='relu', 
        solver='adam', 
        random_state=42,
        verbose=True  # Shows training progress
    )),
])

print("Starting Neural Network Training...")
text_clf.fit(X_train, y_train)
```

```python
predictions = text_clf.predict(X_train)
print("\nTraining Set Performance:")
print(classification_report(y_train, predictions))
```

```python
with open('irb_model.pkl', 'wb') as f:
    pickle.dump(text_clf, f)

print("\nSuccess! Model trained and saved as 'irb_model.pkl'")
```
